```python
# match case statement

number = 2

match number:
    case 1:
        print('One')
    case 2:
        print('Two')
    case 3:
        print('Three')
    case 4:
        print('Four')
    case 5:
        print('Five')
    case _:
        print('othe ')
```

```python
# while loop

i = 1 # Initialization

while i <= 5: # Condition
    print('Data science')
    i = i + 1 # Increment
```

```python
i = 1

while i <= 5:
    print('Data science :', i)
    i = i + 1
```

```python
i = 5

while i >= 1:
    print('Data science :', i)
    i = i - 1 # Decrement
```

```python
i = 1

while i <= 5:
    print('Data science') 
    j = 1
    while j <= 4:
        print('Technology')
        j = j + 1

    i = i + 1
    print()
```

```python
i = 1 

while i <= 5:
    print('Data science', end = ' ') # When we mention end parameter then new line will not create
    j = 1
    while j <= 4:
        print('Technology', end = ' ')
        j = j + 1

    i = i + 1
    print()
```

```python
i = 1 

while i <= 2:
   
    j = 0
    while j <= 2:
        print(i * j, end = ' ')
        j += 1

    i +=1
    print()
```

```python
i = 1

while i <= 5:

    j = 0
    while j <= 4:
        print(i * j, end = ' ')
        j += 1
 
    i += 1
    print()
```

```python
# while else

numbers = [2, 4, 6, 8, 10]

target = 7

i = 0

while i < len(numbers):
    if numbers[i] == target:
        print('Found')
        break
    i = i + 1
else:
    print('Not found')
```
