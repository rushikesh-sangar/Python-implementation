```python
keys = ['name', 'age', 'gender'] # If no value is provided, each key is assigned None.
new_dict = dict.fromkeys(keys)
print(new_dict)
```

```python
keys = ['apple', 'banana', 'cherry'] # A single value is assigned to all keys.
value = 0
inventory = dict.fromkeys(keys, value)
print(inventory)
```
