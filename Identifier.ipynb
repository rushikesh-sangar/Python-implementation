# Python interpreter and compiler

Compiler and interpreter are used to convert a program written in a high level langauge into machine code understood by computers.

Interpreter

1) Translates program one statement at a time.
2) Interpreter run every line item.
3) Execute the single, partial line of code.
4) Easy for programming.
   
Compiler

1) Scans the entire program and translates it as a whole into machine code.
2) No execution if an error occurs.
3) You can not fix the bug (debug) line by line.

3) Special characters are not allowed.
