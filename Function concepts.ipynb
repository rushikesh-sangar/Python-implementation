```python
def greet():
    return 'Good morning'

def add(x, y):
    c = x + y
    return c

def sub(x, y):
    d = x - y
    return d

r = greet()
r1 = add(5, 6)
r2 = sub(5, 6)

print(r)
print(r1)
print(r2)
```

```python
def add_sub(x, y):
    c = x + y
    d = x - y
    return c, d # Return multiple values

result = add_sub(4, 5)
print(result)
print(type(result))
```

```python
def add_sub(x, y):
    c = x + y
    d = x - y
    return c, d # Return multiple values

r, r1 = add_sub(4, 5) 
print(r)
print(r1)
print(type(r))
print(type(r1))
```

```python
def add_sub_mul(x, y):
    c = x + y
    d = x - y
    e = x * y
    return c, d, e

r, r1, r2 = add_sub_mul(4, 5)
print(r)
print(r1)
print(r2)
print(type(r))
print(type(r1))
print(type(r2))
```

## Positional arguments

```python
def person(name, age):
    print(name)
    print(age)
    
person('Nit', 23)
```

```python
def person(name, age):
    print(name)
    print(age)
    
person('Nit')
```

```python
def person(name, age):
    print(name)
    print(age)
    
person('Nit', 23, 234545, 'Hi')
```

```python
def person(name, age):
    print(name)
    print(age)
    
person()
```

```python
def person(name, age):
    print(name)
    print(age)
    
person(23)
```

```python
def person(name, age):
    print(name)
    print(age)
    
person(23, 'Nit')
```

```python
def person(name, age):
    print(name)
    print(age + 1)
    
person(23, 'Nit')
```

## Keyword argument

```python
def person(name, age):
    print(name)
    print(age + 1)

person(age = 23, name = 'Nit')
```

```python
def person(name, age, age2):
    print(name)
    print(age)
    print(age2)

person(age = 20, name = 'Nit', 30)
```

```python
def person(name, age, age2):
    print(name)
    print(age)
    print(age2)

person(age = 20, name = 'Nit', age2 = 30)
```

## Default argument

```python
def person(name, age):
    print(name)
    print(age)

person('Nit')
```

```python
def person(name, age = 18):
    print(name)
    print(age)

person('Nit')
```

## Variabe length arguments (*args)

```python
def sum(a, b):
    c = a + b
    print(c)

sum(5, 6)
```

```python
def sum(a, b):
    c = a + b
    print(c)

sum(5, 6, 7, 8)
```

```python
def sum(a, *b):
    c = a + b
    print(c)

sum(5, 6, 7, 8)
```

```python
def sum(a, *b):
    print(a)
    print(b)

sum(5, 6, 7, 8)
```

```python
def sum(a, *b):
    print(type(a))
    print(type(b))

sum(5, 6, 7, 8)
```

```python
def sum(a, *b):
    c = a
    for i in b: # Fetch each value from the tuple to add them
        c = c + i
    print(c)

sum(5, 6, 7, 8)
```

```python
def add(a, *b):
    c = a
    for i in b:
        c = c + i
    print(c)

sum(5, 6, 7, 8, 10, -1, 100, 1000, 2.3, 199.45)
```

```python
def sum(a, *b):
    c = 0
    for i in b:
        c = c + i
    print(c)

sum(5, 6, 7, 8)
```

## Keyword length arguments (**kwargs)

```python
def person():
    print('Hello')

person('Alex', 36, 'John', 987767)    
```

```python
def person(name, *data):
    print(name)
    print(data)

person('Alex', 36, 'John', 987767)
```

```python
def person(name, *data):
    print(name)
    print(name)

person('Alex', age = 36, home_place = 'South_city', mob = 987767)
```

```python
def person(name, **data):
    print(name)
    print(data)

person('Alex', age = 36, home_place = 'south_city', mob = 987767)
```

```python
def person(name, **data):
    print(name)
    print(data)

person('Alex', age = 36, home_place = 'south_city', mob = 987767, car = 'german')
```
