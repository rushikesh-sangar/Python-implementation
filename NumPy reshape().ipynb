# NumPy reshape()

```python
import numpy as np
```

```python
mydata = np.arange(0, 20)
mydata
```

```python
np.reshape(mydata, (4, 5))
```

```python
mydata
```

```python
mat1 = np.reshape(mydata, (5, 4), order = 'C') # Row major
mat1
```

```python
mat1
```

```python
print(mat1[4, 3])
```

```python
print(mat1[3, 3])
```

```python
mat1
```

```python
print(mat1[-3, -1])
```

```python
mat1
```

```python
mydata
```
