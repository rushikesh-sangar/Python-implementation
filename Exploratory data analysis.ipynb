## Missing value treatment

```python
clean_data
```

```python
clean_data['Age']
```

```python
import numpy as np
```

```python
np.__version__
```

```python
clean_data['Age'] = clean_data['Age'].fillna(np.mean(pd.to_numeric(clean_data['Age'])))
```

```python
clean_data['Age']
```

```python
clean_data['Exp']
```

```python
clean_data['Exp'] = clean_data['Exp'].fillna(np.mean(pd.to_numeric(clean_data['Exp'])))
```

```python
clean_data['Exp']
```

```python
clean_data
```

```python
print(clean_data['Location'].isnull().sum())
```

```python
clean_data['Location']
```

```python
clean_data['Location'].mode()
```

```python
clean_data['Location'] = clean_data['Location'].fillna(clean_data['Location'].mode()[0]) # Selects first mode
```

```python
clean_data['Location']
```

```python
clean_data
```

```python
clean_data.info()
```

```python
clean_data['Age'] = clean_data['Age'].astype(int)
```

```python
clean_data.info()
```

```python
clean_data['Salary'] = clean_data['Salary'].astype(int)
clean_data['Exp'] = clean_data['Exp'].astype(int)
```

```python
clean_data.info()
```

```python
clean_data['Name'] = clean_data['Name'].astype('category')
clean_data['Domain'] = clean_data['Domain'].astype('category')
clean_data['Location'] = clean_data['Location'].astype('category')
```

```python
clean_data.info()
```

```python
clean_data
```
