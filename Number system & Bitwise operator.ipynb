```python
12 | 13
```

## Bitwise Left shift (<<)

```python
12 << 1 # Left shift - It will gain a bit
```

```python
12 << 2
```

```python
bin(25)
```

```python
25 << 1
```

```python
25 << 2
```

## Bitwise Right shift (>>)

```python
12 >> 1 # Right shift - It will lose a bit
```

```python
13 >> 3
```

```python
bin(30)
```

```python
35 >> 1
```

```python
35 >> 2
```
