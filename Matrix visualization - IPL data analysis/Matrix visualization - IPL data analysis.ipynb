# Matrix visualization - IPL data analysis

# Matrix & NumPy

1) Matrix is the tabular representation of data.
2) Lot of data is stored in tabular format & matrix can handle data in tabular format thats why matrix is very important in python.
3) As we are working on dataframes so matrix is very important.
4) List is one dimensional and matrix is multidimensional.
5) Indexation is very important to plot the data points.
6) Here, we will visualize the IPL dataset.
7) Here, we have listed the top 10 highest-paid players for the 2015-2024 seasons.
8) We will analyze how 10 players have been playing over the past 10 years with player data of past 10 years. 
9) Our main goal is to find trends, patterns & their performance for past 10 years.
10) Ultimately they haven't always been top 10 player & lets see how they are imporving what are secreats or patterns.
