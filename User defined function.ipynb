# User defined function

```python
def greet(): # greet() without argument
    print('Good morning')
```

```python
def greet():
print('Good morning')
```

```python
def greet(): # Defining the function
    print('Good morning')
greet() # Calling the function
```

```python
def greet():
    print('Good morning')
greet()

def greet():
    print('Good morning')
greet()
```

```python
def greet():
    print('Good morning')
    print('Good bye')
greet()
greet()
greet()
```

```python
def add(x, y): # x & y are formal arguments
    c = x + y
    print(c)
add(5, 6) # 5 & 6 are actual arguments
```

```python
def add(x, y):
    c = x + y
    print(c)
add(5)
```

```python
def add(x):
    c = x + y
    print(c)
add(5, 6)
```

```python
def add(x, y):
    c = x + y
    print(c)
add(5, 6)
```

```python
def add(x, y):
    c = x + y
    return c
add(5, 6)
```

```python
def add(x, y):
    c = x + y
    print(c) # No value is returned
result = add(5, 6)
print(result)
print(type(result))
```

```python
def add(x, y):
    c = x + y
    return c # Integer value is returned
result = add(5, 6)
print(result)
print(type(result))
```
